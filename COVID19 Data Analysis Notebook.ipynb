```python
# Importing all the modules necessary to create a training set.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Modules are imported')
```

```python
# Take the csv file and run the dataset into the program by 10 rows
covid19_csv_dataset = pd.read_csv("Datasets/covid19_Confirmed_dataset.csv")
covid19_csv_dataset.head(10)
```

```python
covid19_csv_dataset.shape
```

```python
# drop two useless sections from the datasets. Latitude and longitude and
# identify that method as df (short for dataframe)
df = covid19_csv_dataset.drop(["Lat", "Long"], axis=1)
```

```python
df.head(10)
```

```python
# This method groups the dataset by Country and Region. df (dataframe)
# is the dataframe without the latitude and longitude columns.

covid19_aggr_dataset = df.groupby("Country/Region").sum()
```

```python
# this method display all the data information by 10 rows

covid19_aggr_dataset.head(10)
```

```python
# This method is the shape of the dataset

covid19_aggr_dataset.shape
```

```python
# This method will display data of confirmed cases in a specific location 
# e.g. China

covid19_aggr_dataset.loc["China"]
```

```python
# This method will display data of confirmed cases in a specific location 
# e.g. US

covid19_aggr_dataset.loc["US"]
```

```python
# This method will display a plotted graph of confirmed cases in a specific location 
# e.g. China and US

covid19_aggr_dataset.loc["China"].plot()
covid19_aggr_dataset.loc["US"].plot()
```

```python
# This next method will display a plotted graph of confirmed cases in a specific location 
# e.g. China and US

covid19_aggr_dataset.loc["China"][:3].plot()
covid19_aggr_dataset.loc["US"][:3].plot()
```

```python
# This next method will calculate the first derivate of the curve and plot them
# e.g. China

covid19_aggr_dataset.loc["China"].diff()
```

```python
# This next method will calculate the first derivate of the curve and plot them
# e.g. US

covid19_aggr_dataset.loc["US"].diff()
```

```python
# This next method will calculate the first derivate of the curve and plot them
# e.g. China

covid19_aggr_dataset.loc["China"].diff().plot()
covid19_aggr_dataset.loc["US"].diff().plot()
```

```python
# This method will return to us the maximum infection rate for China 

covid19_aggr_dataset.loc["China"].diff().max()
```

```python
# This method will return to us the maximum infection rate for the US

covid19_aggr_dataset.loc["US"].diff().max()
```

```python
# This method will return to us the maximum infection rate for all countries

countries = list(covid19_aggr_dataset.index)
max_inf = []
for c in countries:
    max_inf.append(covid19_aggr_dataset.loc[c].diff().max())
covid19_aggr_dataset["max_inf"] = max_inf
```

```python
covid19_aggr_dataset.head()
```

```python
covid19_data = pd.DataFrame(covid19_aggr_dataset["max_inf"])
```

```python
covid19_data.head()
```

```python
#Create a new report called happiness report and import that csv in the notebook
happy_data = pd.read_csv("Datasets/Worldwide_happiness_report.csv")
happy_data.head()
```

```python
#Have the database remove some columns we do not need
useless_cols = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
happy_data.drop(useless_cols, axis=1, inplace = True)
```

```python
#Call back the updated report 
happy_data.head()
```

```python
#Set the countries as indices instead of the numbers
happy_data.set_index("Country or region", inplace=True)
happy_data.head()
```

```python
happy_data.shape
```

```python
complete_data = covid19_data.join(happy_data, how="inner")
complete_data.head()
```

```python
complete_data.corr()
```

```python
#Plot x as GDP per capita and y as the maximum infection rate...
x = complete_data["GDP per capita"]
y = complete_data["max_inf"]
sns.scatterplot(x, np.log(y))
```

```python
#...to correctly measure the correlation between GDP and maximum infection rate
sns.regplot(x, np.log(y))
```
